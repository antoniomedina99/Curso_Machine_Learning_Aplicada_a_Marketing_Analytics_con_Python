Sub Plots

```python
# Import librerias

import matplotlib.pyplot as plt
# Funci√≥n magica para cargar imagenes 
%matplotlib inline  
import numpy as np      
```

```python
x = np.linspace(0, 100, 101) # crea numero uniformemente espaciados
print(x)
```

```python
y = np.sqrt(x)+ 5 # Se crea con la sqr de "x" y se le suma 5 a cada valor
y
```

```python
plt.figure(figsize= (15,5) ) # Shape
plt.suptitle('Subplots', x=0.4) # la x centra el t√≠tulo

# Plot 1
plt.subplot(1, 3 ,1)
plt.plot(x, y, color = "red")
plt.title("Plot 1")

# plot 2
plt.subplot(1, 3 ,2)
plt.plot(y, x, color = "blue")
plt.title("Plot 2")

# Plot 3
plt.subplot(1,3,3)
plt.plot(x, y, color = "black")
plt.title("Plot 3")

plt.show()
```

```python
plt.figure(figsize= (15,10) ) # Shape
# plt.suptitle('Subplots', x=0.4) # la x centra el t√≠tulo

# Plot 1
plt.subplot(2, 3 ,1)
plt.plot(x, y, color = "red")
plt.title("Plot 1")

# plot 2
plt.subplot(2, 3 ,2)
plt.plot(y, x, color = "blue")
plt.title("Plot 2")

# Plot 3
plt.subplot(2,3,3)
plt.plot(x, y, color = "black")
plt.title("Plot 3")

# Plot 4
plt.subplot(2,3,4)
plt.plot(x, y, color = "orange")
plt.title("Plot 4")

# Ajustar el dise√±o para evitar solapamientos y dejar espacio para el t√≠tulo general
plt.tight_layout()

plt.show()
```
