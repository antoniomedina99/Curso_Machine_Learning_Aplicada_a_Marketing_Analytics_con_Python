# Graficas con el método orientadas a objetos

```python
# Import librerias
import pandas as pd
import matplotlib.pyplot as plt
# Función magica para cargar imagenes 
%matplotlib inline  
import numpy as np
import seaborn as sns

x = np.linspace(0, 100, 101) # crea numero uniformemente espaciados
y = np.sqrt(x)+ 5 # Se crea con la sqr de "x" y se le suma 5 a cada valor
z = np.log(x+ 1)
w = np.cos(y)
```

```python
fig, axes = plt.subplots(2, 3, figsize= (15, 10))
plt.suptitle("Grafica practica subplots")

# plot 1
axes[0,0].scatter(x, y)
axes[0,0].set_title('plot 1 [0,0]')

#plot 2
axes[0,1].scatter(y, x)
axes[0,1].set_title('plot 2 [0,1]')

# plo3
axes[1,1].hist(y)
axes[1,1].set_title('Hist [1,1]')

plt.show()
```

```python
fig, axes = plt.subplots(1,2, figsize= (10, 4))
plt.suptitle('Título de lienzo')

# plot 1
axes[0].scatter(x, y)
axes[0].set_title('Plot 1')

# pot 2
axes[1].hist(y, x)
axes[1].set_title('Hist plot 2')

plt.show()
```

```python
fig, axes = plt.subplots(3,2, figsize = (5, 4))

#plot 1
axes[0,1].hist(y)
axes[0,1].set_title('grafica [0,1]')

#plot 2
axes[2,1].scatter(z, w)
axes[2,1].set_title('plot [2,1] o row3 an colum 2')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1,2, figsize = (10, 5))
fig.suptitle('Canvas del la gráfica')

#plot 1
axes[0].boxplot(x, vert = False)

#plot 2
axes[1].boxplot(x, vert = True)
axes[1].set_title('box plot vertical')

plt.show()
```

# Graficas ingreso Yopal Casanare

```python
# Import data
data_yopal = pd.read_excel("datos modelos yopal.xlsx")
data_yopal.head(3)
```

```python
fig, axes = plt.subplots(2,2)
fig.suptitle('Análisis ingreso Yopal Casanare')

sns.regplot(x = 'edad', y= 'ingreso', data = data_yopal, ax = axes[0,0] )
axes[0,0].grid()

sns.boxplot(x = 'edad', data = data_yopal, ax = axes[0,1], showfliers=True)
axes[0,1].set_title('Edad')
axes[0,1].set_xlabel('valor de x')

axes[0,1].grid()

sns.histplot(data_yopal['edad'], ax= axes[1,0])
axes[1,0].grid()

sns.regplot(x = 'edad', y = 'ingreso_ln', data=data_yopal)
axes[1,1].set_title('Log(ingreso) vs Edad')
axes[1,1].grid()

plt.tight_layout()
plt.show()
```

Nota:
- El ingreso se ajusta mejor cuando se hace una tratnsformación log
- La edad tienen concentración a la izquierda
- La edad mayor a 60 años hacen parto de datos atipicos

```python
# Una grafica con orientación a objetos
fig, axes =  plt.subplots()
sns.regplot(x = 'edad', y = 'ingreso_2', data= data_yopal)
axes.set_xlabel('Edad de los encuestados')
axes.set_ylabel('Cuadrado del ingreso')

plt.grid()
```

```python
# Crear la figura y los ejes utilizando matplotlib de manera correcta
fig, ax = plt.subplots()

# Utilizar seaborn para crear el gráfico de regresión
sns.regplot(x='edad', y='ingreso_2', data=data_yopal, ax=ax)

# Personalización opcional
ax.set_title('Gráfico de Regresión')
ax.set_xlabel('Edad (años)')
ax.set_ylabel('Ingreso')

# Mostrar el gráfico
plt.show()
```

```python
fig, axes = plt.subplots( figsize =(5, 5))
sns.histplot(x = 'edad', data= data_yopal)
fig.suptitle('Histograma de las edades')
axes.set_xlabel('Edad de los encuestados')
axes.set_ylabel('Conteo')
plt.show()
```
