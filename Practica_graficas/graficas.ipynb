Graficas con matplotlib

```python
# Import librerias

import matplotlib.pyplot as plt
# Funci√≥n magica para cargar imagenes 
%matplotlib inline  
import numpy as np                
```

```python
x = np.linspace(0, 100, 101) # crea numero uniformemente espaciados
print(x)
```

```python
y = np.sqrt(x)+ 5 # Se crea con la sqr de "x" y se le suma 5 a cada valor
y
```

```python
plt.plot(x,y, color = "red")
plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')
plt.show()
```
